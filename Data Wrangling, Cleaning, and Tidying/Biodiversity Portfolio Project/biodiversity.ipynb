# Explore the data

### Load the data
Look at the first few row and the columns.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import glob

species = pd.read_csv('species_info.csv')
observations = pd.read_csv('observations.csv')

print(species.head())
print(observations.head())
print('\n')
print(f"species shape: {species.shape}")
print(f"observations shape: {observations.shape}")
```

The species_info.csv contains 4 columns and 5824 rows:
- **category** - The category of taxonomy for each species
- **scientific_name** - The scientific name of each species
- **common_names** - The common names of each species
- **conservation_status** - The species conservation status

The observations.csv contains 3 columns and 23296 rows:
- **scientific_name** - The scientific name of each species
- **park_name** - The name of the national park
- **observations** - The number of observations in the past 7 days

# Tidy the data

### Check for duplicates
Determine how to handle them.

```python
# pd.set_option('display.max_rows', None)
# print(species[species['scientific_name'].duplicated(keep=False)].sort_values(by='scientific_name'))

print('The number of duplicates in species is ' + str(species.duplicated(subset="scientific_name").sum()))
print('The number of duplicates in observations is ' + 
      str(observations.duplicated(subset=["scientific_name", "park_name"]).sum()))

# There should only be one instance of a scientific name (two species cannot have the same scientific name).
# Only one species has different conservation statuses among its duplicates (Canis lupis). 
# drop_duplicates will keep the 'Endangered' Canis lupis.

species = species.drop_duplicates(subset="scientific_name")

# Assuming the data for each species at each park was only supposed to be collected once
observations = observations.groupby(["scientific_name", "park_name"], as_index = False).mean()

# Check to make sure duplicates were dropped
print(species.duplicated(subset="scientific_name").sum())
print(observations.duplicated(subset=["scientific_name", "park_name"]).sum())
```

### Fill in NaN values

```python
species['conservation_status'] = species['conservation_status'].fillna('No Status')
print(species.head())
```

# Analysis

### Let's look at each column

```python
print(f"number of categories: {species['category'].nunique()}\ncategory names: {species['category'].unique()}")
print(f"number of unique species: {species['scientific_name'].nunique()}")
print(f"number of common names: {species['common_names'].nunique()}")
print(f"conservation statuses: {species['conservation_status'].unique()}")
print(f"number of parks: {observations['park_name'].nunique()}\npark names {observations['park_name'].unique()}")
print(f"total observations: {observations['observations'].sum()}")
```

### What is the distribution of conservation_status for animals?

```python
species_nonull = species[species.conservation_status != 'No Status']
green = sns.color_palette("Greens", 4)[-2]
sns.countplot(data=species_nonull, x='conservation_status', hue='conservation_status', \
              palette = sns.color_palette("coolwarm", 4)[:3] + [green], legend=False)
plt.show()

statuses = ['Species of Concern', 'Threatened', 'Endangered', 'In Recovery']
for status in statuses:
    print(f"{status}: " + str((species.conservation_status == status).sum()))
```

```python
statuses = ['Species of Concern', 'Threatened', 'Endangered', 'In Recovery']
for status in statuses:
    status_category = species_nonull[species_nonull.conservation_status==status]
    plt.figure(figsize=(8, 5))
    ax = sns.countplot(data= status_category, x='category', hue='category', palette='Blues', \
                       order=status_category['category'].value_counts(ascending=False).index)
    plt.xticks(rotation=45)
    plt.title(status + " count for each category")
    plt.show()
    counts = status_category.groupby('category').size()
    print(counts)
```

### Next, let's determine if certain types of species more likely to be endangered.

```python
species['protected'] = species.conservation_status != 'No Status'
species['not_protected'] = species.conservation_status == 'No Status'
protection_status = species.groupby(['category'])[['protected', 'not_protected']].sum().reset_index()
protection_status['percent_protected'] = protection_status['protected']/(protection_status['protected']+protection_status['not_protected'])
print(protection_status)
```

It appears that mammals and birds are more likely to be endangered than other animal species. We can now test to see if there is a significant difference in protection status among species.

```python
from scipy.stats import chi2_contingency

# mammals vs birds
contingency = [[30,146],[75,413]]
chi2, pval, dof, expected = chi2_contingency(contingency)
print(pval)

# mammals vs reptiles
contingency = [[30,146],[5,73]]
chi2, pval, dof, expected = chi2_contingency(contingency)
print(pval)
```

A p-value of 0.05 is a good test of significance. Mammals vs birds produces a p-value of 0.69, indicating there is no significant difference between the two species. However, a p-value of 0.038 from a chi2 test between mammals and reptiles represets a significant difference. We can continue to run these tests for each pairing.

### Which species were spotted the most at each park?

```python
# merge the DataFrames
df = pd.merge(species, observations, how='right')
#print(df.shape)
for park in df['park_name'].unique():
    park_animal = df[(df['park_name'] == park) & (df['category'] == 'Mammal')]
    max_value = park_animal.groupby('common_names')['observations'].max().sort_values(ascending=False).reset_index()
    print(max_value.head(5))
    print("The mammal spotted the most at " + park + " was " + max_value.loc[0,'common_names'] + " with " + 
          str(max_value.loc[0, 'observations']) + " observations.")
```

It looks like the Least Shrew, the Eastern Red Bat/Red Bat, Easter Red Bat/Red Bat, and American Mink were the most spotted animals at Bryce, Great Somy Mountains, Yellowstone, and Yosemite Park, respectively. However, we can broaden these observations to just bats, or shrews, or other animals.

```python
# split the common names down to determine the most common type of animal observed
common_names = []
for name in df[df.category == 'Mammal']['common_names']:
    for animal in name.split(','):
        common_names.append(animal.strip())
animal_names = []
for name in common_names:
    for animal in name.split(' '):
        animal_names.append(animal.strip())

animal_counts = pd.DataFrame(animal_names, columns=['names'])
animal_counts = animal_counts['names'].value_counts().reset_index()
print(animal_counts.head())
```

It looks like bats are the most common animal in the species dataset. Let's add a column to the DataFrame so we can use it for our analysis.

```python
df['is_bat'] = df.common_names.str.contains(r"\bBat\b", regex = True)
print(df[df['is_bat']])
```

```python
bat_observations = observations.merge(df[df.is_bat])
bat_observations = bat_observations.groupby(['park_name', 'protected']).observations.sum().reset_index()
bat_observations['observations'] = bat_observations['observations'].astype(int)
print(bat_observations)
```

```python
plt.figure(figsize=(16, 4))
sns.barplot(x=bat_observations.park_name, y= bat_observations.observations, hue=bat_observations.protected)
plt.xlabel('National Parks')
plt.ylabel('Number of Observations')
plt.title('Observations of Bats at Each Park')
plt.show()
```

We can see how bats the protected status of bats at each National Park. We can go back and edit the code to look at the other popular mammals in our species dataset.
